##### Crear un filtro

```python
filtro = df['Director'] == 'Christopher Nolan'
df[filtro] 
```

Filtro con condiciones multiples: operador logico 'y' == '&'

```python
filtro1 = (df['Director'] == 'Christopher Nolan') & (df['Score'] >= 8.5)
df[filtro1]
```

Filtro con condiciones multiples: operador logico 'o' == '|'

```python
filtro2 = (df['Director'] == 'Christopher Nolan') | (df['Score'] >= 8.5)
df[filtro2]
```

Si quieres que devuelva todos los valores **EXCEPTO** los del filtro:

```python
df[-filtro1] # Devuelve todas las filas que no cumplan el filtro
```

Diferencia entre hacer un Filtro, isin() y .contains()

```python
filtro3 = df['Genre'] == 'Drama'
df[filtro3]
```

```python
genero = ['Drama']
filtro4 = df['Genre'].isin(genero)
df.loc[filtro4, 'Genre']
```

```python
filtro5 = df['Genre'].str.contains('Drama')
df.loc[filtro5, 'Genre']
```

##### apply

La función apply itera sobre los elementos de una matriz y aplica una función a cada elemento de la matriz. Por otro lado, el ciclo for es una estructura de control que itera sobre una secuencia de operaciones.

Son iteradores como tal, en el segundo caso con for se puede recorrer una lista, un diccionario, una tupla, y en el primer caso con apply se puede recorrer una lista, un diccionario, una tupla, una matriz.

```python
mayor_que = df['english_title'].apply(len) # Devuelve la longitud de cada título
```

```python
def actualizar_titulo(titulo):
    return titulo.upper()

df['english_title'].apply(actualizar_titulo) # Aqui no se actualiza el dataframe, solo se muestra el resultado
```

```python
df['english_title'] = df['english_title'].apply(actualizar_titulo) # Aqui si se actualiza el dataframe
df
```

##### Añadir y eliminar columnas

¿Como añadir nuevas columnas? En este caso la columna metascore esta en una escala de 0 a 100, mientras que score esta en una escala de 1 a 10

```python
df['total_score'] = ((df['score']) + (df['metascore']/10))/2 # Crea una nueva columna con la media de las puntuaciones
df.head()
```

¿Y para eliminarlas?

```python
df.drop(columns=['total_score'], inplace=True) # Elimina la columna total_score
```

Para eliminar filas con un filtro

```python
df.drop(index = df[df['score'] == 9.3].index) # Elimina la fila que tiene una puntuación de 9.3, falta inplace=True para que se actualice el dataframe
```
