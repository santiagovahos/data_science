## Figuras con graficas simples

````python
import matplotlib.pyplot as plt
import numpy as np
````

### Plot()

````python
x = np.linspace(-10, 10, 100)
y = np.sin(x)
# y_2 = np.cos(x)

plt.figure(figsize = (5, 5))
plt.plot(x, y, marker = "x", color = "blue") # o, s
# plt.plot(x, y_2, marker = "o", color = "red") 
plt.title("Grafica sin(x)")
plt.xlabel("sin(x)")
plt.ylabel("Amplitud");
#plt.show()
````

````python
# Datos
x = np.linspace(-10, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Definir grafica
plt.figure(figsize = (7, 7))
plt.plot(x, y_sin, marker = "x", color = "blue")     # sin
plt.plot(x, y_cos, marker = "x", color = "red")      # cos
plt.plot(x, y_tan, marker = "x", color = "#BF72E6")  # tan

# Personalizar grafica
plt.title("Graficas Trigonométricas")
plt.xlabel("sin(x) --- cos(x) --- tan(x)")
plt.ylabel("Amplitud")
plt.grid(linestyle = "--", linewidth = 1) # color = "grey"
plt.legend(["sin(x)", "cos(x)", "tan(x)"], loc = "upper right") # upper left, lower left
````

````python
# Datos
x = np.linspace(-10, 10, 100)
y_sin = [np.sin(x), "sin(x)", "red"]
y_cos = [np.cos(x), "cos(x)", "blue"]
y_tan = [np.tan(x), "tan(x)", "yellow"]

for y in [y_sin, y_cos, y_tan]:
    
    plt.figure(figsize = (4, 4))
    plt.plot(x, y[0], marker = "x", color = y[2]) 
    plt.title(f"Grafico de {y[1]}")
    plt.xlabel(f"{y[1]}")
    plt.ylabel("Amplitud")
    plt.grid(linestyle = "--", linewidth = 1) # color = "grey"
    plt.legend([f"{y[1]}"], loc = "upper right") # upper left, lower left
````

### Scatter()

Un scatter plot es un grafico de dispersion entre dos variables `una vs la otra`, a continuacion veremos un ejemplo:

````python
# Datos
x = np.arange(0, 100)
y = x**1.5 + np.random.rand(*x.shape)*50

# Definir grafica
plt.figure(figsize = (6, 6))
plt.scatter(x, y)

# Personalizar grafico
plt.title("Grafica de dispersion")
plt.xlabel("Values")
plt.ylabel("Frecuencia")
plt.grid(linestyle = "--", linewidth = 1) # color = "grey"
plt.legend(["Function(x)"], loc = "upper left") # upper left, lower left
````

````python
# Datos
x = np.arange(0, 100)
y = x**1.5 + np.random.rand(*x.shape)*50
y_2 = -x**1.5 + np.random.rand(*x.shape)*50
# Definir grafica
plt.figure(figsize = (6, 6))
plt.scatter(x, y)
plt.scatter(x, y_2)

# Personalizar grafico
plt.title("Grafica de dispersion")
plt.xlabel("Values")
plt.ylabel("Frecuencia")
plt.grid(linestyle = "--", linewidth = 1) # color = "grey"
plt.legend(["Function(x)", "Function(-x)"], loc = "upper left") # upper left, lower left
````

## Figuras compuestas con subplots()

Aunque las graficas anteriores permiten crear visualizaciones simples, la manera mas comun de encontrarnos la forma de crear dichas graficas es mediante el uso de los `subplots`, el cual nos permite hacer lo siguiente en terminos generales:

- Mas informacion: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

<img src = "subplot.png" height = "450" width = "450">

### Graficos

````python
# Datos
x = np.arange(0, 100)
y = x**1.5 + np.random.rand(*x.shape)*50

x_2 = np.linspace(-10, 10, 100)
y_sin = np.sin(x)

fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y)
ax[0, 1].scatter(x, y)
ax[1, 0].plot(x_2, y_sin)
ax[1, 1].plot(x_2, -y_sin)
````
